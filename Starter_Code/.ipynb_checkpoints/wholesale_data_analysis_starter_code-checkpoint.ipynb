## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
data["sub_total"] = data["unit_price"] * data["qty"]
data[["unit_price", "qty", "sub_total"]]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

#data["shipping_price"] = data.apply(lambda row: 7 * row ["unit_weight"] if row["unit_weight"] > 50 else 10 * row["unit_weight"], axis = 1)
#data[["shipping_price","unit_weight", "unit_price", "qty"]].head()
```

```python
def shipping_calculator(weight): 
    if weight > 50: 
         return weight * 7 
    return weight * 10 
#def shipping_calculator(weight):
    #if weight > 50:
        #return weight * 7
    #return weight * 10

data['total_weight'] = data['qty'] * data['unit_weight']
data['shipping_price'] = data['total_weight'].apply(shipping_calculator)
data [['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
data["total_price"] =round ((data["sub_total"] + data ["shipping_price"]) * 1.0925, 2)
data[["total_price", "sub_total", "shipping_price"]].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
data["line_cost"] = data ["unit_cost"] * data ["qty"] + data ["shipping_price"]
# data[["line_cost", "unit_cost", "qty", "shipping_price"]
data.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
data ["profit"] = data ["total_price"] - data ["line_cost"] 
data [["profit", "line_cost", "total_price"]].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_client_spending = data[data["client_id"].isin(top_client_ids)].groupby("client_id")["total_price"].sum()
top_client_spending
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_dataframe = pd.DataFrame ({
    "Total Units Purchased":
    data[data["client_id"].isin(top_client_ids)].groupby("client_id")["qty"].sum(),
        "Total Shipping Price":
    data[data["client_id"].isin(top_client_ids)].groupby("client_id")["shipping_price"].sum(),
        "Total Revenue": "top_client_spending",
        "Total Profit": data[data["client_id"].isin(top_client_ids)].groupby("client_id")["profit"].sum()
})
summary_dataframe
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_dataframe = summary_dataframe.sort_values(by="Total Profit", ascending=False)
summary_dataframe
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
#summary_dataframe["Total Revenue"] /= 1_000_000 
#summary_dataframe["Total Profit"] /= 1_000_000 
summary_dataframe.rename(columns = {
    "Total Units Purchased":  "Total Units Purchased", 
    "Total Shipping Price":  "Total Shipping Price", 
    "Total Revenue (Millions)": "Total Revenue (Millions)", 
    "Total Profit (Millions)": "Total Profit (Millions)"
})
```

```python
summary_dataframe.dtypes
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

money_cols = ["Total Profit"]

def currency_format_millions(amount):
    return f"${amount/1000000:.2f}M"

for col in money_cols:
    summary_dataframe[col] = summary_dataframe[col].apply(currency_format_millions)
```
